```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# **Importing Dataset**


```python
dataset = pd.read_csv('Untitled spreadsheet - Netflix Movie Recommendation Ec.csv',header = None)
transactions = []
for i in range(0,7466):
  transactions.append([str(dataset.values[i,j]) for j in range(0,20)])
```

# **Eclat training on dataset**

```python
from apyori import apriori
rules = apriori(transactions=transactions, min_support=0.003, min_confidence=0.2, min_lift=3, min_length=2)
```

# **Visualizing**

## Raw results

```python
results=list(rules)
```

```python
results
```

## Proper display

```python
def inspect(results):
  movie1 = [tuple(result[2][0][0])[0] for result in results]
  movie2 = [tuple(result[2][0][1])[0] for result in results]
  supports = [result[1] for result in results]
  
  return list(zip(movie1,movie2,supports))
results_indataframe=pd.DataFrame(inspect(results),columns=['Movie1','Movie2','Support'])
```

```python
results_indataframe
```

```python
results_indataframe.nlargest(n=10,columns='Support')
```
